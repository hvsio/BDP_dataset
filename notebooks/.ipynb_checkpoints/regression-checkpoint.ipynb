```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats
from IPython.display import display_html
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../cleaned_data/clean_v4_fillednans.csv', delimiter=',')
df.head(10)
```

```python
corr = df.corr(method='spearman')
_, ax = plt.subplots(figsize=(20,10)) 
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu_r', annot=True, linewidth=0.5, ax=ax)
```

```python
def mosthighlycorrelated(mydataframe, numtoreport, var, second = False):
    # find the correlations
    cormatrix = mydataframe.corr()
    # set the correlations on the diagonal or lower triangle to zero,
    # so they will not be reported as the highest ones:
    cormatrix *= np.tri(*cormatrix.values.shape, k=-1).T
    # find the top n correlations
    cormatrix = cormatrix.stack()
    cormatrix = cormatrix.reindex(cormatrix.abs().sort_values(ascending=False).index).reset_index()
    # assign human-friendly names
    column_name = "FirstVariable" if not second else "SecondVariable"
    cormatrix.columns = ["FirstVariable", "SecondVariable", "Correlation"]
    cormatrix = cormatrix.loc[cormatrix[column_name] == var]
    return pd.DataFrame(cormatrix.head(numtoreport))

def display_two(df1, df2, title1, title2):
    df1_style = df1.style.set_table_attributes('style="display: inline"').set_caption(title1)
    df2_style = df2.style.set_table_attributes('style="display: inline"').set_caption(title2)
    display_html(df1_style._repr_html_()+df2_style._repr_html_(), raw=True)
    

cor_positive = mosthighlycorrelated(df, 10, 'Positive affect')
cor_happiness = mosthighlycorrelated(df, 10, 'Happiness index')
display_two(cor_positive, cor_happiness, 'Correlations with Positive affect', 'Correlations with Happiness index')
```

```python
defaults = [
    'GDP per capita',
    'Social support',
    'Healthy life expectancy at birth',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption',
    'Positive affect',
    'Negative affect'
]

additional_params = [
    'Trust in government',
    'Health spending',
    'Internet access',
    'Home access to computers',
    'Average wage',
    'Poverty rate',
    'Trust in government'
]

questionable_params = [
    'Self-employment rate',
    'Households overcrowding',
    'Self-employment rate'
]

def print_results(model, x, y, xtest, ytest):
    accuracy_test = round(model.score(x, y),4)
    accuracy_train = round(model.score(xtest, ytest),4)
    print(f'Accuracy on training set: {accuracy_train} \nAccuracy on testing set: {accuracy_test}\nModel coeff: {model.coef_}')
    
```

```python
X_original = df[[*defaults]].values
Y = df[['Happiness index']].values
X_train, X_test, Y_train, Y_test = train_test_split(X_original, Y, test_size=0.2, random_state=41)

#Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model_original = LinearRegression()
model_original.fit(X_train, Y_train)
print_results(model_original, X_train, Y_train, X_test, Y_test)
```

```python
X = df[[*defaults, *additional_params, *questionable_params]].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=39)

# Feature scaling
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train, Y_train)
print_results(model, X_train, Y_train, X_test, Y_test) # qusetionable_params increased accuracy on test by 0.5%
```

```python
#training on original values
ols_org = sm.OLS(endog = Y, exog = X_original).fit()
residuals_org = ols_org.resid
pred_org = ols_org.fittedvalues
ols_org.rsquared
```

```python
#training with added values, addition of questionable params increase rsquare by 0.1%
ols = sm.OLS(endog = Y, exog = X).fit()
residuals = ols.resid
pred = ols.fittedvalues
ols.rsquared
```

```python
plt.figure(figsize=(8,6)) 
ax = sns.residplot(pred, residuals)
ax.set_title('Residuals vs Predicted Values')
ax.set_xlabel('Predicted Values')
ax.set_ylabel('Residuals')
plt.show()
```

```python
# PCA to reduce the number of independent variables/reduce dimensionality
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=39)
pca = PCA(n_components=16)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_var = pca.explained_variance_ratio_

model = LinearRegression()
model.fit(X_train, Y_train)
print_results(model, X_train, Y_train, X_test, Y_test)
```
