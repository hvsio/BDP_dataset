```python
import pandas as pd
import numpy as np
```

```python
def clean_data(original_df, cvs_path, variable, year=2020):
    df = pd.read_csv(cvs_path, delimiter=',')[['Entity', 'Year',variable]]
    df = df.loc[df['Year'] == 2017]
    df.rename(columns={'Entity': 'Country'}, inplace=True)
    df_copy = original_df.copy()
    df_copy = pd.merge(df_copy, df, on='Country', how='outer').dropna(thresh=5)
    df_copy.drop(columns=['Year'], axis=1, inplace=True)
    return df_copy
```

```python
main_df = pd.read_csv('../inputs/lifestyle.csv', delimiter=',')

#dropping all city-specific columns
main_df.drop(labels=[
    'Cost of a monthly gym membership(City)',
    'Number of take out places(City)',
    'Cost of a bottle of water(City)',
    'Sunshine hours(City)',
    'Pollution(Index score) (City)',
    'Outdoor activities(City)',
    'Rank'
    ], axis=1, inplace=True)
```

```python
#Load countries to map capitals to countries
country_city_dt = pd.read_csv('../inputs/world-cities_csv copy.csv', delimiter=',')[['name','country']]
country_city_dt.rename(columns={'name': 'City', 'country': 'Country'}, inplace=True)
```

```python
#Add countries
main_df = pd.merge(main_df, country_city_dt, on='City', how='outer')
```

```python
main_df = (main_df
    .dropna(thresh=3)
    .drop_duplicates(subset=['City'])
    .reset_index(drop=True)
)
```

```python
# urban indicator
df_v1 = clean_data(main_df, '../inputs/urban.csv', 'Urban population (% of total population)')
# press freedom indicator
df_v2 = clean_data(df_v1, '../inputs/press-freedom-rsf.csv', 'Press Freedom Score', 2019)
# poverty rates
df_v3 = clean_data(df_v2, '../inputs/poverty.csv', '$1.90 per day - total number of people below poverty line', 2019)
# number of ppl with depression in 2019
df_v4 = clean_data(df_v3, '../inputs/number-with-depression-by-country.csv', 'Prevalence - Depressive disorders - Sex: Both - Age: All Ages (Number)', 2019)
```

```python
# share of 1-person households
df_households = pd.read_csv('../inputs/one-person-households.csv', delimiter=',')[['Entity', 'Year', 'Share of one person households']]
df_households.rename(columns={'Entity': 'Country'}, inplace=True)
df = df_households.groupby('Country').agg({'Year': 'max'})
df = pd.merge(df, df_households, on=['Country', 'Year'], how='left')
df_v5 = pd.merge(df_v4, df, on='Country', how='outer').dropna(thresh=5)
df_v5.drop(columns=['Year'], axis=1, inplace=True)
```

```python
# alcohol consumption
df_alcohol = pd.read_csv('../inputs/alcohol.csv', delimiter=',')[['Entity', 'Year', 'Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)']]
df_alcohol.rename(columns={'Entity': 'Country'}, inplace=True)
df = df_alcohol.groupby('Country').agg({'Year': 'max'})
df = pd.merge(df, df_alcohol, on=['Country', 'Year'], how='left')
df_v6 = pd.merge(df_v5, df, on='Country', how='outer').dropna(thresh=5)
df_v6.drop(columns=['Year'], axis=1, inplace=True)
```
